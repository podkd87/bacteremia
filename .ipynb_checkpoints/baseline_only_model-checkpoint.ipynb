```python
train, valid = train_test_split(baseline,
                                                              train_size=0.7,
                                                              test_size=0.3)
```

```python
train_n, valid_n = train_test_split(new_baseline,
                                                              train_size=0.7,
                                                              test_size=0.3)
```

```python
static_input=Input(shape=(38,),  name='static')
x31 = layers.Dense(16,activation='tanh')(static_input)
x32=layers.BatchNormalization()(x31)
x33=layers.Dropout(0.5)(x32)
#합친 모양
x41 = layers.Dense(16,activation='relu')(x33)
x42=layers.BatchNormalization()(x41)
x43=layers.Dropout(0.5)(x42)
x10=layers.Dense(1, activation='sigmoid')(x43)

model=Model(static_input, x10)
model.compile(loss="binary_crossentropy", optimizer=adam(lr=0.0003), metrics=['accuracy'])
```

```python
path = './only_baseline/'
if not os.path.exists(path):
  os.mkdir(path)

callbacks = [
    ModelCheckpoint(filepath=path+"baseline.hdf5",
                    monitor='val_loss',
                    mode='min',
                    save_best_only=True,
                    save_weights_only=False,
                    verbose=True
    ),
    EarlyStopping(monitor='val_loss', min_delta=0, patience=5, 
                                   verbose=0, mode='auto')]
```

```python
model.fit(train.iloc[:,3:41], train.result, batch_size=500, 
                                        epochs=300, verbose=2, 
                                        validation_data=(valid.iloc[:,3:41], valid.result),
                                        callbacks=callbacks)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import average_precision_score
```

```python
model.load_weights(path+"baseline.hdf5")
```

```python
model_test = model.predict(valid.iloc[:,3:41])
```

```python
roc_val_test = roc_auc_score(valid.result, model_test)
average_precision = average_precision_score(valid.result, model_test)
```

```python
n_estimators_list = [300,1000,3000,5000]

history = []

for n_estimators in n_estimators_list:
    model = LGBMClassifier(boosting_type='gbdt',
                       n_estimators=n_estimators,
                       learning_rate=0.05,
                       num_leaves=31,
                       max_bin=255,
                       min_child_samples=20,
                       min_split_gain=0.0,
                       subsample=0.5,
                       subsample_freq=5,
                       colsample_bytree=0.7,
                       n_jobs=-1,
                       random_state=random_state)
    
    %time model.fit(train_n.iloc[:,4:-1], train_n.result)
     
    y_test_predict = model.predict(valid_n.iloc[:,4:-1])
    
    roc_val_test = roc_auc_score(valid_n.result, y_test_predict)
    average_precision = average_precision_score(valid_n.result, y_test_predict)

    print(f"n_estimators = {n_estimators}, roc = {roc_val_test:.6f},\
    test = {average_precision:.6f}")
    
    history.append({
        'n_estimators': n_estimators,
        'roc_score': roc_val_test,
        'averagePR': average_precision,
    })
    
history = pd.DataFrame(history)
history
```

```python
from sklearn.ensemble import RandomForestClassifier

n_estimators_list = [300,1000,3000]
max_features_list = [0.3,  0.6, 0.8, 1.0]
history = []

for n_estimators in n_estimators_list:
    for max_feature in max_features_list:
        model = RandomForestClassifier(n_estimators = n_estimators,
                                       max_features = max_feature,
                                       class_weight = 'balanced',
                                       random_state=42, n_jobs=-1)

        model.fit(train_n.iloc[:,4:-1], train_n.result)

        y_test_predict = model.predict(valid_n.iloc[:,4:-1])

        roc_val_test = roc_auc_score(valid_n.result, y_test_predict)
        average_precision = average_precision_score(valid_n.result, y_test_predict)

        print(f"n_estimators = {n_estimators}, roc = {roc_val_test:.6f},\
        test = {average_precision:.6f}")

        history.append({
            'n_estimators': n_estimators,
            'max_features':max_feature,
            'roc_score': roc_val_test,
            'averagePR': average_precision,
        })

history = pd.DataFrame(history)
history
```
